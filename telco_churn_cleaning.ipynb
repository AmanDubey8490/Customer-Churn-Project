```python
# Data Cleaning and Preprocessing

# 1) Convert TotalCharges to numeric (invalid -> NaN)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# 2) Drop rows where TotalCharges is NaN (small number ~11 rows)
df = df.dropna(subset=['TotalCharges']).reset_index(drop=True)

# 3) Standardize some service columns that have "No internet service" / "No phone service"
#    Replace them with simple "No" so they can be mapped safely to 0/1.
cols_with_no_internet = [
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]
for c in cols_with_no_internet:
    if c in df.columns:
        df[c] = df[c].replace({'No internet service': 'No'})

# MultipleLines sometimes has "No phone service"
if 'MultipleLines' in df.columns:
    df['MultipleLines'] = df['MultipleLines'].replace({'No phone service': 'No'})

# 4) Now replace Yes/No -> 1/0 for columns that are strictly Yes/No
yes_no_cols = ['Partner', 'Dependents', 'PhoneService', 'PaperlessBilling', 'Churn']
for c in yes_no_cols:
    if c in df.columns:
        df[c] = df[c].replace({'Yes': 1, 'No': 0})

# 5) Also convert the previously standardized service columns
for c in cols_with_no_internet + ['MultipleLines']:
    if c in df.columns:
        df[c] = df[c].replace({'Yes': 1, 'No': 0})

# 6) Optional: convert 'gender' to numeric (Male=1, Female=0) if you want
if 'gender' in df.columns:
    df['gender'] = df['gender'].replace({'Male': 1, 'Female': 0})

# 7) Ensure types are numeric where expected
df['tenure'] = pd.to_numeric(df['tenure'], errors='coerce')
df['MonthlyCharges'] = pd.to_numeric(df['MonthlyCharges'], errors='coerce')
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# 8) Final quick check
print("Shape:", df.shape)
print("\nNon-null counts:")
print(df.info())

# Show top rows to confirm
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1) Make a numeric-only DataFrame (drop ID-like columns first)
if 'customerID' in df.columns:
    df_num = df.drop(columns=['customerID']).select_dtypes(include=['number'])
else:
    df_num = df.select_dtypes(include=['number'])

# 2) Quick check of which numeric columns we have
print("Numeric columns:", df_num.columns.tolist())

# 3) Correlation matrix (only numeric columns)
corr = df_num.corr()

# 4) Show top correlations with Churn
if 'Churn' in corr.columns:
    print("\nTop correlations with Churn:")
    print(corr['Churn'].sort_values(ascending=False))
else:
    print("\nChurn column not numeric / missing in numeric dataframe.")

# 5) Plot heatmap (smaller figure to avoid slow rendering)
plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=False, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap (numeric columns only)")
plt.tight_layout()
plt.show()
```
